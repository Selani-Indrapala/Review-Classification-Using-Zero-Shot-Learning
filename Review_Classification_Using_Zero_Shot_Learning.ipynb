```python
from transformers import pipeline
```

Other models can be found via https://huggingface.co/models?pipeline_tag=zero-shot-classification&sort=downloads

```python
classifier_pipeline = pipeline ("zero-shot-classification", model = "facebook/bart-large-mnli")
```

```python
!pip install datasets
from datasets import load_dataset
```

```python
train_dataset = load_dataset("rotten_tomatoes",split='train')
validation_dataset = load_dataset("rotten_tomatoes",split='validation')
test_dataset = load_dataset("rotten_tomatoes",split='test')
```

```python
n = len(train_dataset['text'])
```

```python
import numpy as np
```

```python
'''input_sequence = train_dataset['text'][40]
label_candidate = ['Good', 'Bad']
classifier_pipeline(input_sequence, label_candidate)'''
#classifier_pipeline(input_sequence, label_candidate,multi_label=True)
```

```python
predictions = np.zeros(n)
for i in range(n):
  input_sequence = train_dataset['text'][i]
  label_candidate = ['Good', 'Bad']
  if (classifier_pipeline(input_sequence, label_candidate))['labels'][0] == 'Good':
    predictions[i] = 1
  else:
    predictions[i] = 0
  print("Case",str(i),"prediction is",str(predictions[i])+". Actual prediction is",str(train_dataset['label'][i]))
```

```python
#Calculating error
x = np.sum(predictions-train_dataset['label'])
print("Error =",str(x/n))
```
